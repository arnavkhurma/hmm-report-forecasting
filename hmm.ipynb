```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# CPI DATA FROM INVESTING.COM

driver = webdriver.Chrome()
driver.get('https://www.investing.com/economic-calendar/cpi-733')

try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'eventHistoryTable733'))
    )
except Exception as e:
    print("Error loading page:", e)
    driver.quit()

try:
    load_more_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, 'showMoreHistory733'))
    )
    while load_more_button.is_displayed():
        load_more_button.click()
        time.sleep(2)
        load_more_button = driver.find_element(By.ID, 'showMoreHistory733')
except Exception as e:
    print("No more content to load or button not found:", e)

soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

table = soup.find('table', id='eventHistoryTable733')

headers = [header.text.strip() for header in table.find_all('th')]

rows = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    rows.append(cols)

# df = pd.DataFrame(rows, columns=headers)
# cpi = df[['Release Date', 'Actual', 'Forecast']]
# cpi.to_csv("CPI-US.csv")   
```

```python
# PPI DATA FROM INVESTING.COM

driver = webdriver.Chrome()
driver.get('https://www.investing.com/economic-calendar/ppi-734')

try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'eventHistoryTable734'))
    )
except Exception as e:
    print("Error loading page:", e)
    driver.quit()

try:
    load_more_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, 'showMoreHistory734'))
    )
    while load_more_button.is_displayed():
        load_more_button.click()
        time.sleep(2)
        load_more_button = driver.find_element(By.ID, 'showMoreHistory734')
except Exception as e:
    print("No more content to load or button not found:", e)

soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

table = soup.find('table', id='eventHistoryTable734')

headers = [header.text.strip() for header in table.find_all('th')]

rows = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    rows.append(cols)

df = pd.DataFrame(rows, columns=headers)
ppi = df[['Release Date', 'Actual', 'Forecast']]
ppi.to_csv("PPI-US.csv") 
```
