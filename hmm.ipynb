```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# CPI DATA FROM INVESTING.COM

driver = webdriver.Chrome()
driver.get('https://www.investing.com/economic-calendar/cpi-733')

try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, 'eventHistoryTable733'))
    )
except Exception as e:
    print("Error loading page:", e)
    driver.quit()

try:
    load_more_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, 'showMoreHistory733'))
    )
    while load_more_button.is_displayed():
        load_more_button.click()
        time.sleep(2)
        load_more_button = driver.find_element(By.ID, 'showMoreHistory733')
except Exception as e:
    print("No more content to load or button not found:", e)

soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

table = soup.find('table', id='eventHistoryTable733')

headers = [header.text.strip() for header in table.find_all('th')]

rows = []
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    rows.append(cols)

# df = pd.DataFrame(rows, columns=headers)
# cpi = df[['Release Date', 'Actual', 'Forecast']]
# cpi.to_csv("CPI-US.csv")   
```

```python
nfp = pd.read_csv('NFP-US.csv')
nfp.drop(columns=["Unnamed: 0"], inplace=True)
nfp
```

```python
nfp['Release Date'] = pd.to_datetime(nfp['Release Date'])
nfp['Actual'] = nfp['Actual'].astype(str).str.replace(',', '', regex=False)
nfp['Actual'] = nfp['Actual'].str.replace('K', '', regex=False).astype(float) * 1000
nfp['Forecast'] = nfp['Forecast'].astype(str).str.replace(',', '', regex=False)
nfp['Forecast'] = nfp['Forecast'].str.replace('K', '', regex=False).astype(float) * 1000
date_range = pd.date_range(start=nfp['Release Date'].min(), end=nfp['Release Date'].max(), freq='D')
nfp = nfp.set_index('Release Date').reindex(date_range, method=None)
nfp = nfp.ffill()
nfp.reset_index(inplace=True)
nfp.rename(columns={'index': 'Date'}, inplace=True)
nfp = nfp[::-1]
nfp.head(20)
```

```python
nfp['Surprise'] = nfp['Actual'] - nfp['Forecast']
```

```python
nfp
```

```python
nfp.to_csv('updated_nfp.csv')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 12))
plt.plot(nfp['Date'], nfp['Surprise'], linestyle='-', label='Surprise')
plt.title('Surprise Over Time - NFP')
plt.xlabel('Date')
plt.ylabel('Surprise')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
